```python
import pandas as pd
import random
from datetime import datetime, timedelta
import os

file_path = "live_sales.csv"

products = ["Laptop", "Phone", "Tablet", "Headphones"]
regions = ["London", "Manchester", "Birmingham", "Leeds"]

random_days = random.randint(0, 10)

data = {
    "OrderID": random.randint(1000, 9999),
    "Date": datetime.now() - timedelta(days=random_days),
    "Product": random.choice(products),
    "Region": random.choice(regions),
    "Sales": random.randint(50, 500),
    "Profit": random.randint(-50, 200)
}

df = pd.DataFrame([data])

if os.path.exists(file_path):
    df.to_csv(file_path, mode='a', header=False, index=False)
else:
    df.to_csv(file_path, index=False)

print("Added:", data)
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("live_sales.csv")
df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

df['Sales'].resample('D').sum().plot()
plt.title("Daily Sales Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Product')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Product")
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Region")
plt.show()
```
